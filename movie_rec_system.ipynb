```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
credits.shape
```

```python
## merge base on title
movies = movies.merge(credits,on = 'title')
```

```python
movies.head()
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies = movies[['id','title','overview','genres','keywords','cast_x','crew_x']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
## in cast  take top 3
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python

movies['cast_x'] = movies['cast_x'].apply(convert3)
movies.head()
```

```python
movies['cast_x'] = movies['cast_x'].apply(lambda x:x[0:3])
```

```python

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew_x'] = movies['crew_x'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
# remove space between names
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python

movies['cast_x'] = movies['cast_x'].apply(collapse)
movies['crew_x'] = movies['crew_x'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.sample(4)
```

```python

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tagss'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast_x'] + movies['crew_x']
```

```python
new_movies = movies.drop(columns=['overview','genres','keywords','cast_x','crew_x'])
```

```python
new_movies['tagss'] = new_movies['tagss'].apply(lambda x: " ".join(x))
```

```python
new_movies.head()
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movies['tagss']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Superman')
```
